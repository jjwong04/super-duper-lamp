```python
import pandas as pd 
product_list = [['AIG', 'TravelWise', 'Deluxe'], ['Allianz', 'Single Trip', 'Gold'], ['AIG', 'TravelWise', 'Standard'], ['Allianz', 'Single Trip', 'Silver'], ['AIG', 'TravelWise', 'Basic'], ['Allianz', 'Single Trip', 'Bronze'] ] 

df_product = pd.DataFrame(product_list, columns=['Insurer', 'Product', 'Plan']) 

score_list = [['Medical Score', 0.3, 7.96, 6.56, 6.04, 7.27, 6.52, 9.83], ['Personal Effects Score', 0.3, 9.73, 8.68, 6.39, 7.61, 9.2, 7.5], ['Trip Delay Score', 0.2, 8.94, 6.85, 8.78, 8.96, 7.09, 9.09], ['Trip Cancellation Score', 0.2, 6.03, 6.76, 8.3, 8.45, 6.77, 8.87] ] 

df_score = pd.DataFrame(score_list, columns=['Detailed Score', 'Weighting', 'AIG: TravelWise - Deluxe', 'AIG: TravelWise - Standard', 'AIG: TravelWise - Basic', 'Allianz: Single Trip - Gold', 'Allianz: Single Trip - Silver', 'Allianz: Single Trip - Bronze'])
```

```python
df_product['Full Name']=df_product['Insurer']+": "+df_product['Product']+" - "+df_product['Plan']
df_product.head()
```

```python
df_score['Weighting'] = pd.to_numeric(df_score['Weighting'])
total_sums = [0] * 6
overall_score = []
for i in range(len(df_score)):
    for j in range(6):
        overall_score = df_score.iloc[i, 1] * df_score.iloc[i, j + 2]
        total_sums[j] += overall_score
        
new_row = pd.DataFrame([total_sums], columns=df_score.columns[2:])
df_score=pd.concat([df_score, new_row], ignore_index=True)
```

```python
column_names = df_score.columns.tolist()

for j in range(6):
    lookup_row = df_product.loc[j, 'Full Name']
    for i in range(6):
        lookup_column = column_names[i+2]
        if lookup_column == lookup_row:
            df_product.loc[j, 5] = df_score.iloc[4, i+2]

df_product=df_product.rename(columns={5: 'Overall Score'})
```
